```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib as mpl
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('PotatoPrice.csv')
print(df)
```

```python
mpl.__version__
```

```python
#visualization of data in graph using matplotlib
%matplotlib inline

xlb=plt.xlabel('potato in kg')
ylb=plt.ylabel('potato price')
plt.style.use('classic')
plt.scatter(df.potato_kg, df.price) 
```

```python
x=df[['potato_kg']]

y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(x,y,test_size=0.2)
```

```python
print(X_train)
```

```python
#Model creation of ANN

model=tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=[1],activation='linear')])
```

```python
model.summary()
```

```python
#Compile the ANN model
model.compile(optimizer='sgd',
             loss='mean_squared_error',metrics=['accuracy'])
```

```python
#training the ANN model

model.fit(X_train,y_train, epochs=500)
```

```python
model.predict(X_test)
```

```python
test_loss, test_acc=model.evaluate(X_test,y_test,verbose=2)
```
